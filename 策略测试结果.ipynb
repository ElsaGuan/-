所以本part将关注小市值股票，针对小市值股票开发交易策略

为了减少小市值股票在这样的熊市中的风险，我研究了这一期间策略的归因分析报告（由聚宽平台提供）。

上图是风格因子分析的结果。

如图所示，HML因子为正，表示这一策略的收益很大一部分依赖于对高价值股票的投资。

根据这一信息，在选取的小市值股票中，再加一个选股因子来选取小市值股票中的高价值股票进行投资。

在这里，选取账面价格比（BP ratio）作为高价值股票的衡量标准：即账面价格比越高，说明股票为搞价值股票。

策略思路：
1. 股票池为上证指数成分股
2. 股票池中选取10分位的小市值股
3. 小市值股中根据账面价值比率由高到低排序，选取前十的股票进行交易
4. 调仓周期为一个月
5. 测试期间是 2013-06-27 ~ 2018-01-04，覆盖一个牛熊周期
